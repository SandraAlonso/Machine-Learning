### Observaciones:

 ## Análisis de componentes principales (APC)

```python
X = data.iloc[:,2:23] #Seleccionamos todas las columnas quitando la columna Unnamed (id de la fila ) y la columna Target.
pca = PCA(n_components=21)
pca.fit(X)
# En este array se muestra el porcentaje de variación de cada uno de los componentes.
# Como podemos obseverar no hay apenas variación.
print(pca.explained_variance_ratio_)
```
